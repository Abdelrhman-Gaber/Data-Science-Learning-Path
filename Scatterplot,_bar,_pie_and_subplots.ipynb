# Scatterplot, bar, pie and subplots

```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

## scatterplot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.scatter(x,y,marker='o')

 
plt.xlabel("time")
plt.ylabel("value")
plt.title("sine wave")
```

## bar plot




```python
data = {'Apple':25, 'Banana':15, 'Cherry':20, 'Mango':35}
fruits = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (10, 5))

plt.bar(fruits, values, color ='green',width = 0.4)
 
plt.xlabel("Fruit name")
plt.ylabel("No. of fruits")
plt.title("number of fruits")
```

## pie chart

```python
plt.pie(values,labels=fruits)
```

## subplots

```python
z=np.cos(x)
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.xlabel("time")
plt.ylabel("value")
plt.title("sine wave")
plt.subplot(1, 2, 2)
plt.plot(x,z)
plt.xlabel("time")
plt.ylabel("value")
plt.title("cosine wave")
```
